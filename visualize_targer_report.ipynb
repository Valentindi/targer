```python
fp_to_report = "targer-xlnet-base-capri-patence=100-rnn-hidden-dim=200-epochs=10000-lr_bert0.005-frozen=TRUE.txt"
```

```python
#fp.readlines()
```

```python
fp = open(fp_to_report, "r")
epoch, train_loss, f1_macro_train, f1_macro_dev, f1_macro_test = [], [], [], [], []
best_f1_macro_vals_train, best_f1_macro_vals_test, best_f1_macro_vals_dev = [], [], []
best_f1_macro_train, best_f1_macro_test, best_f1_macro_dev = 0, 0, 0

line_with_data = False
for it, line in enumerate(fp.readlines()):
    if "-------------------------------------------------------------------------------------" in line:
        if line_with_data:
            break
        else:
            line_with_data = True
            continue
    if not line_with_data: continue
        
    values = [float(x) for x in line.split("|")]
    epoch.append(values[0])
    train_loss.append(values[1])
    f1_macro_train.append(values[2])
    f1_macro_dev.append(values[3])
    f1_macro_test.append(values[4])
    
    if best_f1_macro_train < values[2]:
        best_f1_macro_train = values[2]
    best_f1_macro_vals_train.append(best_f1_macro_train)
    
    if best_f1_macro_test < values[3]:
        best_f1_macro_test = values[3]
    best_f1_macro_vals_test.append(best_f1_macro_test)
    
    if best_f1_macro_dev < values[4]:
        best_f1_macro_dev = values[4]
    best_f1_macro_vals_dev.append(best_f1_macro_dev)
    
```

```python
import pandas as pd
df = pd.DataFrame()
df["epoch"] = epoch
df["loss"] = train_loss
df["f1_macro_train"] = f1_macro_train
df["f1_macro_test"] = f1_macro_test
df["f1_macro_dev"] = f1_macro_dev

df["best_f1_macro_train"] = best_f1_macro_vals_train
df["best_f1_macro_test"] = best_f1_macro_vals_test
df["best_f1_macro_dev"] = best_f1_macro_vals_dev
```

```python
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline

fig = plt.figure(figsize=(20, 16))

plt.title('report ' + fp_to_report)  # Add a title so we know which it is

plt.plot(df["epoch"], df["loss"], label='loss')
plt.plot(df["epoch"], df["f1_macro_train"], label='train')
plt.plot(df["epoch"], df["f1_macro_test"], label='test')
plt.plot(df["epoch"], df["f1_macro_dev"], label='dev')
plt.legend()
```

```python
plt.savefig(fp_to_report + ".png")
```

```python
%matplotlib inline

fig = plt.figure(figsize=(20, 16))

plt.title('best values ' + fp_to_report)  # Add a title so we know which it is

plt.plot(df["epoch"], df["loss"], label='loss')
plt.plot(df["epoch"], df["best_f1_macro_train"], label='train')
plt.plot(df["epoch"], df["best_f1_macro_test"], label='test')
plt.plot(df["epoch"], df["best_f1_macro_dev"], label='dev')
plt.legend()
```
